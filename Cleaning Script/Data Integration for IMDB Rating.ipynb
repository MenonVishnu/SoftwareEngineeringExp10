```python
import pandas as pd
```

```python
movies = pd.read_csv('.././DATA/ml-25m/ml-25m/movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('.././DATA/ml-25m/ml-25m/ratings.csv')
ratings.head()
```

```python
ratings.count()
```

```python
groupratings = ratings.groupby('movieId').rating.mean()
groupratings.head()
```

```python
groupratings.count()
```

```python
result = pd.merge(movies, groupratings, on='movieId')
result.tail(20)
```

```python
for index, movie in enumerate(result['title']):
    movie = movie.split()
    movie.pop()
    movie1 = ' '
    movie = movie1.join(movie)
#     print(movie)
    result['title'][index] = movie
result['title']
```

```python
result_copy = result
result_copy.head()
```

```python

result.head()
```

```python
result.to_csv('ratings1.csv', index=False)
```
