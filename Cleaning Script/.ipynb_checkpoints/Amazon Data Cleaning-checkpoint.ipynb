# Amazon Prime Video Data Cleaning

```python
import pandas as pd
```

## Importing Data

```python
amazon_data = pd.read_csv('.././DATA/amazon prime movies.csv')
amazon_data.head(30)
```

## Format Conversion of time from hh:mm to mm

```python
amazon_data_copy = amazon_data
for index,i in enumerate(amazon_data_copy['Running Time']):
    if 'h' in i:
        if len(i.split()) == 1:
            i = i.split('h')
            i = int(i[0])*60
            amazon_data_copy['Running Time'][index] = i
            continue
        i = i.split()
        if (len(i) > 2):
            i = int(i[0])*60+int(i[2])
            amazon_data_copy['Running Time'][index] = i
        continue
    
    elif 'min' in i:
        i = int(i.split('m')[0])
        amazon_data_copy['Running Time'][index] = i

    
amazon_data = amazon_data_copy      
        
```

```python
amazon_data['platform'] = 'Amazon Prime'
```

```python
amazon_data.drop(['Plot'], axis=1,inplace=True)
```

## Droped Rows with more than 3 Null Values
