```python
amazon_data = pd.read_csv('.././Cleaning Script/Amazon Merged Data.csv')
```

## Changing Age to different formats

```python
for index,i in enumerate(amazon_data['Age']):
    if(amazon_data['type'][index]=='TV Show'):
        if(i=='18+'):
            amazon_data['Age'][index] = 'TV-MA'
        elif(i=='16+' or i=='13+'):
            amazon_data['Age'][index] = 'TV-14'
        elif(i=='7+'):
            amazon_data['Age'][index] = 'TV-PG'
        else:
            amazon_data['Age'][index] = 'TV-G'
    else:
        if(i=='18+'):
            amazon_data['Age'][index] = 'R'
        elif(i=='16+' or i=='13+'):
            amazon_data['Age'][index] = 'PG-13'
        elif(i=='7+'):
            amazon_data['Age'][index] = 'PG'
        else:
            amazon_data['Age'][index] = 'G'
amazon_data.head()
```

```python
amazon_data.head()
```

## Removing unnecesary genres 

```python
for index,i in enumerate(amazon_data['genres']):
    if(len(i)==1):
        continue;
    else:
        i = i.split(',')
        amazon_data['genres'][index] = i[0]
amazon_data.head()
```

## Changing \N to N/A in running time

```python
for index,i in enumerate(amazon_data['Running Time']):
    if(i=="\\N"):
        amazon_data['Running Time'][index] = 'N/A'
        
amazon_data.head()
```

```python
amazon_data.to_csv('Amazon_Cleaned.csv',index=False)
```
